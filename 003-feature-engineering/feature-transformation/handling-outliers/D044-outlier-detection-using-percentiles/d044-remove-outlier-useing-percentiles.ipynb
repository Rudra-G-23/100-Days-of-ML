# Capping --> Winsorization

```python
# Capping --> Winsorization
df['Height'] = np.where(df['Height'] >= upper_limit,
        upper_limit,
        np.where(df['Height'] <= lower_limit,
        lower_limit,
        df['Height']))
```

```python
df['Height'].describe()
```

```python
fig = plt.figure(figsize=(17, 5))

ax1 = plt.subplot(131)
sns.histplot(df['Height'], kde=True, color='blue')

ax2 = plt.subplot(132)
sns.boxenplot(df['Height'], color='blue')

ax3 = plt.subplot(133)
sns.boxplot(df['Height'], color='blue')

plt.show()
```
