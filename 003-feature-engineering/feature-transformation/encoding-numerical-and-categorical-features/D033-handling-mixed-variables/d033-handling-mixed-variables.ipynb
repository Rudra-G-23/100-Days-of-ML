# import data and libraries

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic.csv')
df.sample(5)
```

```python
df['number'].unique()
```

```python
fig = df['number'].value_counts().plot(kind='bar')
fig.set_title("Passengers traveling with family")
```

# (Number Column) Extract the numerical & categorical parts 

```python
# Extract the numerical parts 
df['number_numerical'] = pd.to_numeric(df['number'], errors='coerce', downcast='integer')
df['number_numerical'] 
```

```python
df['number_numerical'].value_counts().plot(kind='bar')
```

```python
# Extract the categorical parts 
# np.where(condition, true, false)

df['number_categorical'] =  np.where(
    df['number_numerical'].isnull(),
    df['number'],
    np.nan
)

df['number_categorical'].unique()
```

```python
df['number_categorical'].value_counts()
```

# (Cabin Column) Extract the numerical and categorical data

```python
df['Cabin'].unique()
```

```python
df['Ticket'].unique()
```

```python
# Captures numerical part
df['cabin_num'] = df['Cabin'].str.extract('(\d+)') 
df['cabin_num']
```

```python
df['cabin_num'].value_counts().plot(kind='bar')
```

```python
# Captures the first letter
df['cabin_cat'] = df['Cabin'].str[0]
df['cabin_cat']
```

```python
df['cabin_cat'].value_counts().plot(kind='barh')
```

# ( Ticket Column) Extract the numerical and categorical data

```python
df['Ticket'].apply(lambda x: x.split())
```

```python
# extract the last bit of ticket as number
df['ticket_num'] = df['Ticket'].apply(lambda x: x.split()[-1])

# In this here we can found some alphabets
# so we need more refine
df['ticket_num'].unique() 
```

```python
df['ticket_num'] = pd.to_numeric(
    df['ticket_num'],
    errors='coerce',
    downcast='integer'
)

df['ticket_num']
```

```python
# extract the first part of ticket as category
df['ticket_cat'] = df['Ticket'].apply(lambda x: x.split()[0])
df['ticket_cat']
```

```python
# more filters
df['ticket_cat'] = np.where(
    df['ticket_cat'].str.isdigit(),
    np.nan,
    df['ticket_cat']
)

df['ticket_cat']
```

```python
df['ticket_cat'].unique()
```

```python
df['ticket_cat'].value_counts().plot(kind='pie')
```
