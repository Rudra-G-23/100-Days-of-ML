# Ordinal Encoding for categorical variables

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
df['education'].unique()
```

```python
df['review'].unique()
```

- In Ordinal Encoding we know that 
  
  **In Review Column**
- Poor < Average < Good 
- Poor -> 0
- Average -> 1
- Good -> 2

  **In Review Column**
- School < UG < PG 
- school -> 0
- UG -> 1
- PG -> 2

According to the rank we need to write in same order.
